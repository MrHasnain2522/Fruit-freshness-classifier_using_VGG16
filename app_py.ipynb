```python
# Import libraries
import gradio as gr
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from PIL import Image

# Load model
model = load_model("/content/fruits_fresh_rotten (1).keras")

# Clases define
class_names = [
    'Fresh Apple',
    'Fresh Banana',
    'Fresh Orange',
    'Rotten Apple',
    'Rotten Banana',
    'Rotten Orange'
]

# Prediction function
def predict(img):
    if isinstance(img, np.ndarray):
        img = Image.fromarray(img.astype('uint8'), 'RGB')  # Convert to PIL

    img = img.resize((256, 256))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0

    preds = model.predict(img_array)[0]
    return {class_names[i]: float(preds[i]) for i in range(len(class_names))}

# Launch Gradio interface
gr.Interface(
    fn=predict,
    inputs=gr.Image(type="numpy"),
    outputs=gr.Label(num_top_classes=6),
    title="Fruit Fresh vs Rotten Classifier",
    description="Upload an image of a fruit to classify it as fresh or rotten,Ensure images will be apple,orange,and bnanna."
).launch(share=True)


```
